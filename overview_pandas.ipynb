## grouping

```python
data_5 = {'company' : ['toyota' , 'lamborgini' ,'lamborgini', 'buggti' , 'buggti' , 'toyota' , 'hyunda'] ,
          'car' : [ 'gt86' , 'urus'  , 'aventador', 'shiron' , 'viron' , 'supra', 'jenesis'] ,
          'top speed' : [220 , 290 ,315, 310 ,315 , 240 , 260] ,
          'engine' :[4 , 8 ,12, 12 ,12 , 6 , 6] ,
          'horse power' : [350 ,570 , 700 , 810 , 750 , 525 , 470]}

dataframe_3 = pd.DataFrame(data_5)

C0 = dataframe_3.groupby(['company' ,'top speed']).max()




print(f'dataframe 3 :\n{dataframe_3}')
print("--------------------------")
print(f'C0 : \n{C0}')
print("--------------------------")



```
